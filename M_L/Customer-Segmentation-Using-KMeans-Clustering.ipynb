```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Mall_Customers.csv')
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
df.drop(['CustomerID'],axis=1, inplace=True)
```

```python
X = df.iloc[:,[2,3]].values
```

## Choosing the number of Clusters 
#### WCSS -> with in cluster sum of square

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i, n_init=10, random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
sns.set()
plt.plot(range(1,11), wcss)
plt.title('Elbow graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

### Optimum number of cluster=5 because it goes continious after point 5 elbow in comparision of 3

```python
# Now train KMeans for 5
kmean = KMeans(n_clusters=5, n_init=10, random_state=0)
y = kmean.fit_predict(X)
print(y)
```

```python
# Plotting clusters
plt.figure(figsize=(8,8))
plt.scatter(X[y==0,0], X[y==0,1], s=50, c='green', label='Cluster 1')
plt.scatter(X[y==1,0], X[y==1,1], s=50, c='red', label='Cluster 2')
plt.scatter(X[y==2,0], X[y==2,1], s=50, c='blue', label='Cluster 3')
plt.scatter(X[y==3,0], X[y==3,1], s=50, c='violet', label='Cluster 4')
plt.scatter(X[y==4,0], X[y==4,1], s=50, c='pink', label='Cluster 5')

plt.scatter(kmean.cluster_centers_[:,0], kmean.cluster_centers_[:,1], s=100, color='black', label='Centroid')
plt.title('Customer Groups')
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.show()
```
