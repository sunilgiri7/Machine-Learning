## count or Frequency Encoding

```python
import pandas as pd
df = pd.read_csv('adult.csv',usecols=['workclass','education','marital-status','occupation','relationship','race','gender','native-country'])
df.head()
```

```python
for feature in df.columns[:]:
    print(feature,":",len(df[feature].unique()),'labels')
```

```python
df['native-country'].value_counts()       #value_counts().to_dict()
```

```python
country = df['native-country']
```

```python
country = pd.get_dummies(country,drop_first = True)
country.head()
```

```python
df.head()
```

```python
import numpy as np
df['sex'] = np.where(df['gender'].isnull(),1,0)
df.head()
```

```python
df.drop(['gender'],axis=1,inplace=True)
df.head()
```

```python
df = df.join(country)
df.head()
```

```python
df.drop(['native-country'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df['relationship'].dropna()
```

```python
relation = df['relationship']
relation = pd.get_dummies(relation,drop_first=True)
relation
```

```python
df.drop(['relationship'],axis=1,inplace=True)
df = df.join(relation)
df.head()
```

```python
df['race'].value_counts()
```

```python
race = df['race']
race = pd.get_dummies(race,drop_first=True)
race.head()
```

```python
df.drop(['race'],axis=1,inplace=True)
df.head()
```

```python
df = df.join(race)
df.head()
```

```python
def impute_nan(df, variable):
    df[variable+"newvar"]=np.where(df[variable].isnull(),"Missing",df[variable])
for features in ['occupation']:
    impute_nan(df,features)
df.head()
```

# Count/Frequency Encoding

```python
# Now let's replace each label in occupation by its count

# first we make a dict that maps each labels to counts 
df_frequency = df.occupation.value_counts().to_dict()
```

```python
df.occupation.head()
```

```python
df.occupation = df.occupation.map(df_frequency)
df.head()
```

```python
df['marital-status'].value_counts()
```

```python
status = df['marital-status']
status = pd.get_dummies(status,drop_first=True)
status.head()
```

```python
df.drop(['marital-status'],axis=1,inplace=True)
df = df.join(status)
df.head()
```

```python
frequency = df.education.value_counts().to_dict()
```

```python
df.education = df.education.map(frequency)
df.head()
```

```python
df['workclass'] = df['workclass'].str.replace('?','local_public')
df.head()
```

```python
df.workclass.value_counts().to_dict()
```

```python
class_map = {'Private': 8,
             'Self-emp-not-inc': 7,
             'Local-gov': 6,
             'local_public': 5,
             'State-gov': 4,
             'Self-emp-inc': 3,
             'Federal-gov': 2,
             'Without-pay': 1,
             'Never-worked': 0
}
df['workclass_ordinal'] = df.workclass.map(class_map)
df.head()
```

```python
df.drop(['workclass'],axis=1,inplace=True)
df.head()
```
