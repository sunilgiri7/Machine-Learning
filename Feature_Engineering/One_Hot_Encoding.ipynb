```python
import pandas as pd
import numpy as np
data = 'https://raw.githubusercontent.com/krishnaik06/Complete-Feature-Engineering/master/mercedesbenz.csv'
```

```python
df = pd.read_csv(data,usecols=['X1','X2','X3','X4','X5','X6'])
df.head(5)
```

```python
#lets have a look at how many labels each variable has
for col in df.columns:
    print(col,':',len(df[col].unique()),'labels')
```

```python
# kets eximine how many columns we will obtain after one hot encoding these variables
pd.get_dummies(df,drop_first=True).shape
```

```python
# let's find the top 10 most frequent categories for the variable x2
df.X2.value_counts().sort_values(ascending=False).head(20)
```

```python
# let's make a list with the most frequent categories of the variable
top = [x for x in df.X2.value_counts().sort_values(ascending=False).head(10).index]
top
```

```python
for label in top:
    df[label] = np.where(df['X2']==label,1,0)
    
df[['X2']+top].head(40) 
```
